```python
from getdist import plots
import getdist 
```

```python
sample23 = getdist.mcsamples.loadMCSamples("./data/F24-") 
sample26 = getdist.mcsamples.loadMCSamples("./data/F27-") 
```

```python
# 3D (scatter) triangle plot
g = plots.get_subplot_plotter(width_inch=6)
g.triangle_plot([sample26, sample23], 
                ['m0', 'm12', 'A0', 'tanb', 'gminus2','omega'], 
                plot_3d_with_param='mh', 
                legend_labels=['Packages', 'Expressions'])
g.export('1and2.pdf')
```
